3.3.1.4.2.3 Análise para  remoção e criação das novas features

a - ClasseImovel -> quantidade de andares

```python
# Demonstração da distribuição da classe de imóvel
sns.countplot(dft['ClasseImovel'])
```

```python
# Subdividir em duas categorias - 1o piso e 2o piso - modelo 1
# dft['ClasseImovel_X'] = np.where((dft['ClasseImovel'] <= 120) | (dft['ClasseImovel'] == 150), 0, 1)
# apresentou erro, usei outra forma
#KeyError: "None of [Index(['ClassImovel_2'], dtype='object')] are in the [columns]"
```

```python
# Subdividir em duas categorias - 1o piso e 2o piso - modelo 2
conditions = [
    (dft['ClasseImovel'].isna()),
    (dft['ClasseImovel'] <= 50) | (dft['ClasseImovel'] == 120) | (dft['ClasseImovel'] == 150),
    (dft['ClasseImovel'] > 50) & (dft['ClasseImovel'] < 120) | (dft['ClasseImovel'] == 160)  | (dft['ClasseImovel'] == 190),
      ]
values = [np.nan,1,2]
dft['ClassImovel_2'] = np.select(conditions, values) #Construindo a nova variável
dft[['ClassImovel_2']].value_counts(sort = False) #Avaliando as frequencias
```

b - Condição Imóvel

```python
dft[dft['CondicaoImovel'] == 0]
```

```python

# Demonstração da distribuição da condicao do imóvel
sns.countplot(dft['Condicao'])
```

```python
# Subdividir em 4  categorias 1 - Ruim (1 a 4 ) / 2 - Médio ( 5 ) / 3 - Acima da média (6, 7 e 8) / 4 - Excelente (9 e 10)
conditions = [
    (dft['Condicao'].isna()),
    (dft['Condicao'] <= 4),
    (dft['Condicao'] == 5),
    (dft['Condicao'] >= 6) & (dft['Condicao'] <= 8),
    (dft['Condicao'] == 9) | (dft['Condicao'] == 10),
      ]
values = [np.nan,1,2,3,4]
dft['CondicaoImovel'] = np.select(conditions, values) #Construindo a nova variável
dft[['CondicaoImovel']].value_counts(sort = False) #Avaliando as frequencias
```

```python
# Demonstração da distribuição da nova feature CondicaoImovel
sns.countplot(dft['CondicaoImovel'])
```

c - Valor do Metro Quadrado- ValorMetroQd

```python
# Criar feature com o valor do metro quadrado por bairro
# Dividir o valor do imóvel pela área construída por bairro
dft['ValorMetroQd'] = dft['PrecoVenda'] / dft['AreaConstruida']
dft['ValorMetroQd'].head()
```

d - Categoria por Bairro - CatBairro

```python
# d - Criar feature com a categoria do bairo
# Verifica frequencia dos bairro
dft['Bairro'].value_counts()[:10]
```

```python
# Identificação dos bairros com preços maiores
dft.groupby(['Bairro'])['ValorMetroQd'].nlargest().head(10)
```

```python
# Encontrar a mediana do valor do mt2 por bairro
dft_bairro = dft.groupby(['Bairro'])['ValorMetroQd'].agg('mean')
```

```python
# Criar dataframe com bairro e valor do mt2
dfb = dft_bairro.to_frame()
dfb
```

```python
dfb['ValorMetroQd'] = dfb['ValorMetroQd'].round(0)
dfb
```

```python
# Converte valor mt2 para int64
dfb['ValorMetroQd'] = dfb.astype('int64')
dfb.sort_values("ValorMetroQd", ascending=False)
dfb.info()
```

```python
# Cria a categoria CatBairro
# Subdividir em 5  classes 1,2,3,4 e 5
conditions = [
    (dft['ValorMetroQd'].isna()),
    (dft['ValorMetroQd'] > 140),
    (dft['ValorMetroQd'] > 125) & (dft['ValorMetroQd'] < 140),
    (dft['ValorMetroQd'] > 118) & (dft['ValorMetroQd'] < 125),
    (dft['ValorMetroQd'] > 104) & (dft['ValorMetroQd'] < 118),
    (dft['ValorMetroQd'] < 104)]      
values = [np.nan,1,2,3,4,5]
dft['CatBairro'] = np.select(conditions, values) #Construindo a nova variável
dft[['CatBairro']].value_counts(sort = False) #Avaliando as frequencias
```
